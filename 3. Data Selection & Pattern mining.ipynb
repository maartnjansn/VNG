```python
import pandas as pd
import numpy as np
import support as sp
```

```python
data = pd.read_csv('data/out/df.csv')

drop = ['Unnamed: 0', 'Unnamed: 0.1', 'Causal change', 'Time change', 'Ttl panels', 'Ttl pages', 'Sort field',
        'Genre', 'Country', 'Pub', 'Year', 'Title', 'Magazine Title', 'Drawer', 'Issue/Chapter', 'Story title', 'Writer', 
        'Study', 'Panels/page', 'Macro', 'Mono', 'Micro', 'Amorphic', 'Amb', 'Characters', 'Spatial location', 'Causal change', 
        'Time change']
checker = ['character_change', 'spatial_change', 'causal_change', 'time_change']

for name in checker:
    buf = pd.Series(data[name])
    data[name] = pd.to_numeric(buf, errors='coerce')
    
temp = pd.Series(data['macro'])
data['macro'] = pd.to_numeric(temp, errors='coerce')
temp2 = pd.Series(data['coded_page'])
data['coded_page'] = pd.to_numeric(temp2, errors='coerce')

data = data.drop(drop, axis=1, errors='ignore')
data = data.rename(str.lower, axis='columns')
data = data.rename(index = str, columns = {'original year' : 'original_year'})

# CHECK IF AMB == AMBIGUOUS
#data.to_csv('data/out/data.csv')
#print(len(np.unique(data['title'])))
```

```python
data = data.loc[:,['id', 'title', 'decade', 'continent', 'country_notype', 'country_decade', 'coded_page', 'panel_nr', 'macro','mono', 
                   'micro', 'amorphic', 'amb', 'character_change', 'spatial_change', 'causal_change', 'time_change']]

seq_d = {}
l = []
checker = ['character_change', 'spatial_change', 'causal_change', 'time_change']
# create dict with format [title, decade, continent, (sequence of panels e.g. macro,mono,macro,mono,micro)]
for i in range(len(data)-1): ## !!!! check indexing at the end of the file
    # find column that contains 1 and append to list, creating a sequence of panels
        if data.id[i] == data.id[i+1]:
            #print(data.id[i], data.id[i+1], data.title[i], data.title[i+1])
            for col in data.iloc[:,-9:]:
                if col not in checker:
                    if data[col][i] > 0:
                        l.append(col)
                else:
                    continue
                    if data[col][i+1] > 0:
                        l.append(col)
            # Append the sequence to the dict and start with next comic
        else:
            for col in data.iloc[:,-5:]:
                if data[col][i] > 0:
                    l.append(col)
            l.extend((data.decade[i], data.continent[i], data.country_notype[i], data.country_decade[i]))
            seq_d[data.id[i]] = l
            l = []
                
print(len(seq_d))
```

```python
sequence = []
seq_asia = []
seq_america = []
seq_europe = []
def data_transform_split(split=False):
    if split:
        for i in seq_d:
            if seq_d[i][-3] == 'Asia':
                seq_asia.append(seq_d[i])
            elif seq_d[i][-3] == 'America':
                seq_america.append(seq_d[i])
            else:
                seq_europe.append(seq_d[i])
                
        return seq_asia, seq_america, seq_europe
    else:
        for i in seq_d:
            sequence.append(seq_d[i])
            
        return sequence

sequence = data_transform_split()
print(sequence)
```
