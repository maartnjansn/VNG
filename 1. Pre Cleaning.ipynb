```python
# write frequency of headers to dictionary and clean as much as possible
d = {}

for items in headers:
    for entry in items:
        temp = items.values[0][entry]
        temp = re.sub(r'[^a-zA-Z\s]','',temp)
        temp = temp.lower().rstrip()
        #temp = re.sub(r'[^a-zA-Z\s]','',temp)
        
       # try:
        #    if temp[-1] == " ":
         #       temp = temp[:-1]
       # except:
        #    pass
        
        temp = re.sub(r"\s+", '_', temp)
        if temp in d:
            d[temp] += 1
        else:
            d[temp] = 1
            
print(d)
#print(c)
```

```python
# Further cleaning and merging of columnnames

for item in list(d):
    if item == 'macromono' or item == 'macromacro':
        del d[item]
    if item == 'divsional_montage':
        d['divisional'] += d[item]
        del d[item]
    if item == 'coded_page':
        del d[item]
    if item == 'actual_page':
        d['page'] += d[item]
        del d[item]
    if item == '':
        d['title'] += d['']
        del d[item]
    if item == 'spatial_change':
        d['spatial_location'] += d[item]
        del d[item]
    if item == 'temporal_change':
        d['time_change'] += d[item]
        del d[item]
    if item == 'character_change':
        d[item] += d['characters']
        del d['characters']
    if item == 'base':
        d['base_framing'] += d[item]
        del d[item]
    if item == 'publisher':
        d[item] += d['pub']
        del d['pub']
    if item == "mediumclose":
        d['med_close'] += d[item]
        del d[item]
    if item == 'subj':
        d['subjective_viewpoint'] += d[item]
        del d[item]
    if item == 'chapter':
        d['issuechapter'] += d[item]
        del d[item]
    if item == 'year':
        del d[item]
```
